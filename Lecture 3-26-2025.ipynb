```python
my_text_list[2].replace("?", "")
```

```python
re.sub(r"[A-Z]+", "_", my_text_list[2]) # r at the begining of a python string before "" means 'raw string' (interpret the string inside as a literal), the + means that any character inside the bracket one or more times (looks at the next one and keeps going until something not inside the brackets)
```

```python
re.sub(r"[A-Z]*", "_", my_text_list[2]) # * means 0 or more times, replace -- It's also looking at the empty string, so that's why you get so many '_'s
```
