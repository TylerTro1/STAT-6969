# README

This project spawned out of the interest in Assignment 5. I wanted to try new pre-processing methods, new models, and run cross validation for different hyperparameters. Once I tried to implement all of that manually, I found it highly tedious and figured it would be worthwhile to make a general function which searched over all combinations of pre-processing(s), model(s), and hyperparameter(s), returning the best combination on the given dataset (that is, yielding the best accuracy / F1-Score). 

For this final project, please first look through all of the attached .py files. Proceed in the following order: 

model.py
* Majority Baseline -- Basic Model which returns the majority class in a multi-class classifier
* Support Vector Machine -- Support Vector Machine model of my own creation which uses stochastic gradient descent
* Logistic Regression -- Logistic Regression Model of my own creation which uses stochastic gradient descent

utils.py
* clip -- truncates large and small values. Used in model.py
* shuffle_data -- shuffles a dataset randomly. Used for epoch evaluation in model.py

train.py
* init_model -- just instantiates a model with the given hyperparameters

evaluate.py
* accuracy -- returns the proportion of correct labels given prediction labels and true labels


cross_validation.py
* cross_validation -- performs a cross product search for the best hyperparameters if given folds of data

final_project.py
* Tries a bunch of pre-processing methods, manually, on the loaded dataset. May also be performed on other numerical data if loaded manually. 
* Measures and attempts to make evaluation claims about different model types using accuracy() and F1-Score

final_project_functions.py
* Attempts to generalize all of the manual work done in final_project.py to solve the general notion of "which combination of pre-processing, hyperparameters and models achieve the best results on a given dataset?"
* Inspired by a Google research project that does essentially this

main.py
* to be run
