```python
def load_one_file(filename):
    image = Image.open(filename)
    resized_image = image.resize((25,25))
    image_array = np.array(resized_image)
    flattened_array = image_array.reshape(-1)
    return flattened_array
```

```python
def load_one_file_without_new_variable_names(filename):
    image = Image.open(filename)
```

```python
def load_one_file_really_compactly(filename):
    return np.array(Image.open(filename).resize((25,25))).reshape(-1)
```
