### Numpy and Pandas

```python
# if you do not have numpy and pandas installed, first do so with:
# pip install numpy
# pip install pandas
# or, if you are using conda:
# conda install numpy
# conda install pandas
import numpy as np
import pandas as pd
```

```python
# convert the lists to numpy arrays and add the elements respectively
list_1 = [1, 2, 3, 4, 5]
list_2 = [22, 71, 14, 0, 45]

np_list_1 = np.array(list_1)
np_list_2 = np.array(list_2)

print(np_list_1 + np_list_2)
```

```python
# what other operations can you do with these numpy arrays?
np_list_1 * np_list_2
np_list_1 / np_list_2
np_list_1 - np_list_2
```

```python
# convert the elements of one of the arrays to floats

float_np_list_1 = np_list_1.astype(np.float32)
```

```python
# find the sum and mean of the elements of list_2

sum = np.sum(np_list_2)
print(sum)

mean = np.mean(np_list_2)
print(mean)
```

```python
# make a numpy array of all integers from 1 to 100 and turn it into a 10x10 square array

int_list = np.arange(1, 101)
int_matrix = int_list.reshape(10,10)
print(int_matrix)
```

```python
# multiply the 10x10 array with itself using component-wise multiplication

int_matrix ** 2
```

```python
# multiply the 10x10 array with itself using standard matrix multiplication

int_matrix.dot(int_matrix)
```

```python
# create three arrays of length 100 of random numbers of any distribution you choose

one = np.random.uniform(size = 100)
two = np.random.normal(0, 1, 100)
three = np.random.exponential(1, 100)
```

```python
# create a pandas DataFrame with the arrays you just created as columns
# set the names of the columns to X1, X2, and X3
d = {'X1': one, 'X2': two, 'X3': three}
data = pd.DataFrame(d)
print(data)
```

```python
# calculate the mean of all columns of the data frame at once

col_means = data.mean()
print(col_means)
```

```python
# do the same with the sum, and the standard deviation

col_sums = data.sum()
print("Sums: ")
print(col_sums)
col_std = data.std()
print("\nStandard Deviations: ")
print(col_std)
```

```python
# use the describe method to see all key statistical summaries of all the columns of the data frame

data.describe()
```

```python
# create a new column X4 which is in some way a combination of any of the first three columns
data['X4'] = data['X1']*data['X3'] - data['X2']
print(data)
```
