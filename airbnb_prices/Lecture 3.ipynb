What to do with missing data? 
* Figure out information sought/needed
* Imputing - eg, replace missing values with the mean
* drop the missing data

```python
# Drops a row if any column is missing for that row--this will prove to be useless
albany.dropna(how = "any")
```

```python
# Drops a row if every column is missing for that row
albany.dropna(how = "all").shape
```

```python
# drops any column that is missing at least one value in one row
albany.dropna(axis="columns", how="any").shape
```

```python
# drops any column that is missing all data for all rows in that column
albany.dropna(axis = "columns", how = "all").shape
```
