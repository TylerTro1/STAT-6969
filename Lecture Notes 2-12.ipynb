Two major types of machine learning algorithms: 
* Supervised Learning (labeled data)
* Unsupervised Learning (unlabeled data)

Among supervised learning, there are two major types of problems: 
* Classification Problem (discrete target variable -- typically binary)
* Regression (continuous target variable)

```python
# For this problem, since the target variable is binary (M or B), we'll use classification algorithms to predict it
y = cancer_data.diagnosis
x = cancer_data.drop(columns = ['diagnosis', 'Unnamed: 32', 'id'])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
cancer_scalar = StandardScaler()
```

```python
cancer_scalar.fit(x_train)
```

```python
x_train_scaled = cancer_scalar.transform(x_train)
x_test_scaled = cancer_scalar.transform(x_test)
```

```python
preprocessed_cancer_logreg = LogisticRegression()
```

```python
preprocessed_cancer_logreg.fit(x_train_scaled, y_train)
```

```python
preprocessed_cancer_logreg.score(x_test_scaled, y_test)
```

NOTE: We just improved the accuracy by pre-processing\\
We scaled in this case

```python
# might clip outliers
# scale/standardize features
# imputing (fill in or delete missing data, say with the mean)
```
