2.1 How many congressional districts are there in each state?

```python
%%sql
SELECT s.name AS state, COUNT(DISTINCT cd.name) AS num_districts
FROM states AS s
JOIN congressional_districts AS cd ON s.state = cd.state
GROUP BY s.name
```

2.2 What is the average number of black poeple per congressional district in Florida?


Solution took me a bit of research since it bothered me the table wasn't ordered by district number. My original query was: 
* %%sql
* SELECT cd.name AS district, AVG(100.0 * cd.black / cd.total_population) AS percent_black
* FROM congressional_districts AS cd
* WHERE cd.name LIKE '%Florida%'
* GROUP BY cd.name

```python
%%sql
SELECT 
    cd.name AS district, 
    AVG(100.0 * cd.black / cd.total_population) AS percent_black
FROM congressional_districts AS cd
WHERE cd.name LIKE '%Florida%'
GROUP BY cd.name
ORDER BY 
    CAST(
        SUBSTR(
            cd.name,
            INSTR(cd.name, 'District ') + 9,
            INSTR(SUBSTR(cd.name, INSTR(cd.name, 'District ') + 9), ' ') - 1
        )
    AS INTEGER)
```

2.3 What is the total number of households who live in a "place" that has a name containing the letters "q" or "x"?

```python
%%sql
SELECT SUM(p.total_households) AS total_households_in_place_with_p_or_x
FROM places AS p
WHERE p.name LIKE '%p%' OR p.name LIKE '%x%'
```

```python
%%sql
SELECT DISTINCT(p.place) AS place, p.name AS location, p.total_households
FROM places AS p
WHERE p.name LIKE '%x%' OR p.name LIKE'%p%'
```

2.4 What is the total income for each state? 

```python
%%sql
SELECT COUNT(s.state)
FROM states AS s
```

The extra two states are 'Puerto Rico' and 'District of Columbia', so we remove those. Note: The state number exceeds 50, but that is because some numbers are skipped (such as 07). I also looked up how to put it into a standard numerical format with a $-sign. 

```python
%%sql
SELECT s.name AS name, printf("$%, d", s.total_population * s.per_capita_income) AS formatted_state_income
FROM states AS s
WHERE s.name != 'Puerto Rico' AND s.name != 'District of Columbia'
ORDER BY s.total_population * s.per_capita_income DESC
```
