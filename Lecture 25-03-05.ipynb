```python
%%sql
SELECT *
FROM states
JOIN congressional_districts
ON states.state = congressional_districts.state
LIMIT 5
```

Types of joins:
* INNER JOIN (only rows that are found in both tables)
* LEFT JOIN (all rows from left table, even if no match in right table)
* RIGHT JOIN (all rows from right table, even if no match from left table)
* OUTER JOIN (all rows from both tables regardless of match)
* CROSS JOIN (all possible ways to match rows from left and right tables--result is len(left table) * len(right table))

```python
%%sql
SELECT *
FROM states AS s
JOIN congressional_districts AS cd
ON s.state = cd.state
LIMIT 5
```

```python
%%sql
SELECT s.name, COUNT(*) AS num_of_districts
FROM states AS s
JOIN congressional_districts AS cd
ON s.state = cd.state
GROUP BY s.name
```
