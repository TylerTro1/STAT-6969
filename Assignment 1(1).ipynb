```python
# define a function that takes two integers as input with the following behavior
# if both numbers are positive, return the sum of the two
# if either number is negative, return the product of the two
# raise a ValueError otherwise

def my_func(x, y):
    if isinstance(x, int) == False or type(y) != int:
        raise ValueError("Either x or y is not an integer")
    if x > 0 and y > 0:
        return x + y
    if x < 0 or y < 0:
        return x*y
    else:
        raise ValueError("Either x or y was equal to zero, while the other was positive or zero")

print(my_func(1, 2))
print(my_func(-1, -2))
print(my_func(-1, 2))
# my_func(0, 0) 
# my_func(0, 1)
print(my_func(-2, 0))
# my_func("hi", 2)
my_func(True, 2)
```

```python
# write a for loop that prints out the contents of the dictionary in the form "key: <key>, value: <value>"
my_dict = {
    "sam": 2,
    "jill": 5,
    "bombadil": "ten",
    4: "eighty-two"
}

for key in my_dict:
    print(f"key: {key}, value: {my_dict[key]}")
```
