```python
def remove_stopwords(input):
    tokenized_input = input.split()
    filtered_input = [token for token in tokenized_input if token not in ENGLISH_STOP_WORDS]
    rejoined_output = " ".join(filtered_input)
    return rejoined_output
```

```python
tweets["filtered_text"] = tweets.text.apply(remove_stopwords)
```
