```python
%%sql
SELECT name,
    CAST(white AS FLOAT) / total_population AS ratio_white,
    CAST(black AS FLOAT) / total_population AS ratio_black,
    CAST(hispanic AS FLOAT) / total_population AS ratio_hispanic,
    CAST(asian AS FLOAT) / total_population AS ratio_asian,
    CAST(american_indian AS FLOAT) / total_population AS ratio_american_indian,
    CAST( pacific_islander AS FLOAT) / total_population AS ratio_pacific_islander,
    CAST(other_race AS FLOAT) / total_population AS ratio_other_race
FROM states
LIMIT 10
```

```python
%%sql
-- want to order by highest to lowest ratio of hispanic population
SELECT name,
    CAST(white AS FLOAT) / total_population AS ratio_white,
    CAST(black AS FLOAT) / total_population AS ratio_black,
    CAST(hispanic AS FLOAT) / total_population AS ratio_hispanic,
    CAST(asian AS FLOAT) / total_population AS ratio_asian,
    CAST(american_indian AS FLOAT) / total_population AS ratio_american_indian,
    CAST( pacific_islander AS FLOAT) / total_population AS ratio_pacific_islander,
    CAST(other_race AS FLOAT) / total_population AS ratio_other_race
FROM states
ORDER BY ratio_hispanic DESC
LIMIT 10
```

```python
# For each state, find the lowest median_household_income district within that state
```

```python
%%sql
SELECT s.name, MIN(cd.median_household_income) AS min_income
FROM congressional_districts cd
JOIN states s ON s.state = cd.state
GROUP BY s.name
LIMIT 5
```

```python
%%sql 
SELECT MAX(min_income_district)
FROM(SELECT s.name, MIN(cd.median_household_income) AS min_income_district
FROM congressional_districts cd
JOIN states s ON s.state = cd.state
GROUP BY s.name
LIMIT 5)
```

```python
%%sql 
SELECT name, min_income_district
FROM(SELECT s.name, MIN(cd.median_household_income) AS min_income_district,
row_number() OVER (ORDER BY MIN(cd.median_household_income) DESC) AS row_num
FROM congressional_districts cd
JOIN states s ON s.state = cd.state
GROUP BY s.name
LIMIT 5)
WHERE row_num = 1
```

```python
%%sql
SELECT s.name, MIN(cd.median_household_income) AS min_income_district
FROM congressional_districts cd
JOIN states s ON s.state = cd.state
GROUP BY s.name
ORDER BY min_income_district DESC
LIMIT 1
```
